# finding and Remove outliers:-

```python
#percentail and IQR method to detect outlier

q1=np.percentile(data['price'],25)
q3=np.percentile(data['price'],75)
print('25 and 75 percentile',q1,q3)
IQR=q3-q1
IQR

LC_outlier = q1-1.5*IQR
UC_outlier = q3+1.5*IQR
print(LC_outlier,UC_outlier)
data.loc[(data['price']>UC_outlier)&(data['price']<LC_outlier)]
```

## ML models:-

```python
## ML models:-

''' 
1). Select the target/predict variable  
2). split the data into train and test
3). apply the ML model based on the problem.
4). Fit on training data.
5). predict on test data.

'''
```

```python
y=data['price']
y
```

```python
x=data.drop(columns='price', axis=1)
x=data.drop(columns='date',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lr = LinearRegression()
```

```python
Lr.fit(x_train, y_train)
```

```python
Lr_Predct = Lr.predict(x_test)
Lr_Predct
```

## Interpretation:-

```python
print(Lr.intercept_)
```

```python
Lr.coef_
```

```python
df_coef=pd.DataFrame(Lr.coef_, x.columns, columns=['coefficient'])
df_coef
```

## Model evaluate:

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE=mean_absolute_error (y_test,Lr_Predct)
MAE
```

```python
MSE=mean_squared_error(y_test,Lr_Predct)
MSE
```

```python
RMSE=np.sqrt(MSE)
RMSE
```
